```python
import pandas as pd
from scipy.stats import wilcoxon
```

```python
cad_raw_data = pd.read_csv('cad_dados_mnp.csv',sep=',')
```

```python
cad_raw_data=cad_raw_data.drop(columns='val_renda_seguro_desemp_memb')
cad_raw_data['val_outras_rendas_memb'] =cad_raw_data['val_outras_rendas_memb'].fillna(0)
```

```python
from scipy.stats import binom_test
sucessos = (cad_raw_data['cod_sabe_ler_escrever_memb'] == 1).sum()
n = 10000
#Probabilidade esperada sob a hipótese nula
p_esperado = 0.952

# Teste binomial unilateral (greater) para H1: p > 0.952
resultado = binom_test(sucessos, n, p_esperado, alternative='greater')

print(f"Número de sucessos (sabe ler e escrever): {sucessos}")
print(f"Tamanho da amostra: {n}")
print(f"Valor-p do teste binomial: {resultado}")

# Comparação com alfa = 0.10
if resultado < 0.10:
    print("Rejeitamos a hipótese nula. A proporção de indivíduos que sabem ler e escrever é significativamente maior que 95.2%.")
else:
    print("Não rejeitamos a hipótese nula. Não há evidência suficiente para afirmar que a proporção de indivíduos que sabem ler e escrever é maior que 95.2%.")
```

```python


dados = cad_raw_data[cad_raw_data['val_renda_bruta_12_meses_memb'] >=0]['val_renda_bruta_12_meses_memb'].to_numpy()
# Mediana esperada da população, ajustada para o teste
mediana_esperada = 380*12  # Exemplo, ajuste conforme necessário

# Calcula a diferença entre cada valor dos dados e a mediana esperada
diferencas = dados - mediana_esperada

# Aplicar o teste de Wilcoxon (alternativa para testar medianas) com alternative='greater'
# Considera apenas as diferenças não nulas para o teste
stat, p_value = wilcoxon(diferencas[diferencas != 0], alternative='greater')

print(f"Valor-p do teste de quantis (Wilcoxon): {p_value}")

# Comparação com alfa = 0.10
if p_value < 0.10:
    print("Rejeitamos a hipótese nula. A mediana dos dados é significativamente maior que a mediana esperada.")
else:
    print("Não rejeitamos a hipótese nula. Não há evidência suficiente para afirmar que a mediana dos dados é maior que a mediana esperada.")
```

```python
from scipy.stats import chisquare
observados = [7472, 2528]  # Exemplo: 9472 sabem, 2528 não sabem, total 1000

# Frequências esperadas sob a hipótese nula de 95.2% saberem ler e escrever
total = sum(observados)
esperados = [0.952 * total, 0.048 * total]

# Realiza o teste qui-quadrado
chi2_stat, p_value = chisquare(f_obs=observados, f_exp=esperados)

print(f"Estatística qui-quadrado: {chi2_stat}")
print(f"Valor-p do teste qui-quadrado: {p_value}")

# Comparação com alfa = 0.10
if p_value < 0.10:
    print("Rejeitamos a hipótese nula. A proporção de indivíduos que sabem ler e escrever é significativamente diferente de 95.2%.")
else:
    print("Não rejeitamos a hipótese nula. Não há evidência suficiente para afirmar que a proporção é diferente de 95.2%.")
```

```python
dados = cad_raw_data['cod_sabe_ler_escrever_memb'].replace(2, 0).replace(1, 1).dropna().to_numpy() 

sucessos = sum(dados)  # Contando o número de '1's
n = len(dados)  # Tamanho da amostra

# Proporção esperada
p_esperado = 0.952  # Exemplo: proporção de sucesso esperada (mediana teórica)

# Realizando o teste binomial
resultado = binom_test(sucessos, n, p_esperado, alternative='two-sided')

print(f"Número de sucessos: {sucessos}")
print(f"Tamanho da amostra: {n}")
print(f"Valor-p do teste binomial (teste de quantil): {resultado}")

# Conclusão do teste
alfa = 0.10
if resultado < alfa:
    print("Rejeitamos a hipótese nula. A mediana/proporção observada é significativamente diferente do valor esperado.")
else:
    print("Não rejeitamos a hipótese nula. Não há evidência suficiente para afirmar que a mediana/proporção é diferente do valor esperado.")
```

```python
# Codificando a variável 'val_renda_bruta_12_meses_memb' em 1 (acima de 4500) e 0 (abaixo de 4500)
cad_raw_data['renda_codificada'] = cad_raw_data['val_renda_bruta_12_meses_memb'].apply(lambda x: 1 if x > 4500 else 0)

# Contagem de sucessos (renda > 4500) e tamanho da amostra
sucessos = cad_raw_data['renda_codificada'].sum()  # Contando os valores 1
n = len(cad_raw_data['renda_codificada'])  # Tamanho total da amostra

# Defina a proporção esperada, por exemplo, 10% dos indivíduos com renda acima de 4500
p_esperado = 0.10  # Ajuste conforme a proporção esperada

# Realizando o teste binomial
resultado = binom_test(sucessos, n, p_esperado, alternative='two-sided')

print(f"Número de sucessos (renda > 4500): {sucessos}")
print(f"Tamanho da amostra: {n}")
print(f"Valor-p do teste binomial (teste de quantil adaptado): {resultado}")

# Comparação com alfa = 0.10
alfa = 0.10
if resultado < alfa:
    print("Rejeitamos a hipótese nula. A proporção de renda > 4500 é significativamente diferente da proporção esperada.")
else:
    print("Não rejeitamos a hipótese nula. Não há evidência suficiente para afirmar que a proporção de renda > 4500 é diferente da proporção esperada.")
```
